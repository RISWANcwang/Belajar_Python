Terakhir, program mencetak seluruh isi list yang telah diperbarui setelah semua operasi selesai dilakukan.

```python
mahasiswa = ["Arya", "Aril", "Andika", "Ishaq", "Rizal"]
mahasiswa.append("Hasana")
mahasiswa.insert(3, "Kolbi")
mahasiswa.pop()
mahasiswa.sort(reverse=True)
mahasiswa.remove("Arya")

i = 1
for nama in mahasiswa:
    print("Nama Mahasiswa ke-", i ,": ", nama)
    i+=1
    
print(mahasiswa)
```

#### Analisis

Kode tersebut melakukan beberapa operasi untuk memanipulasi list bernama mahasiswa. Pada bagian perulangan for, setiap nama mahasiswa ditampilkan bersama nomor urutnya dengan bantuan variabel penghitung i, dan akhirnya program mencetak isi list yang telah diperbarui serta diurutkan setelah semua proses selesai dilakukan.

```python
person = {"name": "Riswan", "hp": "12345", "kelas": "TEKOM F"}
print(person)
```

#### Analisis

Kode tersebut mendefinisikan sebuah dictionary bernama person yang berisi tiga pasangan key-value, yaitu "name" dengan nilai "Riswan", "hp" dengan nilai "12345", dan "kelas" dengan nilai "TEKOM F".
```python
person = {"name": "Riswan", "hp": "12345", "kelas": "TEKOM F"}
for key, value in person.items():
    print(key, ":", value)
```

```python
i = 1
while i <= 9:
    print(i)
    i += 2
```

#### Analisis

Kode tersebut menggunakan perulangan while untuk menampilkan deretan angka dari 1 hingga 9 dengan kenaikan sebesar 2 pada setiap iterasi. Variabel i dimulai dengan nilai 1, lalu setiap kali perulangan dijalankan nilainya bertambah 2 menggunakan i += 2. Selama kondisi i <= 9 terpenuhi, program akan terus mencetak nilai i. Hasil akhirnya, program menampilkan deretan angka 1, 3, 5, 7, dan 9, dan perulangan berhenti ketika nilai i melebihi 9.

```python
i = 1
while i <= 13:
    if i == 7:
        print("WOW")
    else:
        print(i)
    i +=2
```

#### Analisis

Kode tersebut menggunakan perulangan while yang berjalan dari nilai 1 hingga 13 dengan kenaikan sebesar 2 pada setiap iterasi. Di dalam perulangan, program memeriksa apakah nilai i sama dengan 7 menggunakan pernyataan if. Jika kondisi tersebut terpenuhi, program mencetak "WOW", sedangkan jika tidak, program mencetak nilai i seperti biasa. Akibatnya, output yang dihasilkan adalah 1, 3, 5, WOW, 9, 11, dan 13, dan perulangan berhenti ketika nilai i melebihi 13.

```python
for i in range(1, 11):
    if i <= 8:
        if i % 2 == 1:
            print("hidup")
        else:
            print("mahasiswa")
    else:
        print(i)
```

#### Analisis

Kode tersebut menggunakan perulangan for dengan range(1, 11) untuk menjalankan iterasi dari angka 1 hingga 10. Di dalam perulangan terdapat percabangan bersarang (nested if) yang memeriksa beberapa kondisi. Jika nilai i kurang dari atau sama dengan 8, maka program akan mengecek apakah i merupakan bilangan ganjil atau genap. Jika i ganjil (i % 2 == 1), maka akan dicetak "hidup", sedangkan jika i genap, akan dicetak "mahasiswa". Untuk nilai i yang lebih besar dari 8 (yaitu 9 dan 10), program langsung mencetak nilai i tanpa pemeriksaan tambahan.

```python
for i in range(1, 14):
    print(i)
    if i % 3 == 2:  
        print("Ha")
```

#### Analisis

Kode tersebut menggunakan perulangan for dengan range(1, 14) untuk menghasilkan deretan angka dari 1 hingga 13. Pada setiap iterasi, program akan mencetak nilai i terlebih dahulu, kemudian memeriksa apakah i % 3 == 2 (yakni ketika i dibagi 3 menghasilkan sisa 2). Jika kondisi ini benar, maka program akan mencetak kata "Ha" setelah angka tersebut. Dengan demikian, output yang dihasilkan menampilkan angka 1 hingga 13, di mana setelah angka 2, 5, 8, dan 11 akan muncul tulisan "Ha".

```python
for i in range(1, 14):
    if i not in [3, 5, 6, 9, 10, 12]:
        print(i)
        
    if i in [2, 5, 8, 11]:
        print("Ha")

    if i in [4, 9]:
        print("Hi")
```

#### Analisis

Kode tersebut menggunakan perulangan for dari 1 hingga 13 untuk menampilkan kombinasi antara angka dan kata tertentu berdasarkan kondisi yang telah ditentukan. Jika nilai i tidak termasuk dalam daftar [3, 5, 6, 9, 10, 12], maka program akan mencetak nilai angka tersebut. Selanjutnya, apabila i termasuk dalam [2, 5, 8, 11], program akan mencetak kata "Ha", dan jika i termasuk dalam [4, 9], maka akan mencetak kata "Hi". Dengan demikian, output yang dihasilkan merupakan urutan angka dan kata yang bergantian sesuai kondisi yang terpenuhi dalam setiap iterasi.
