## Global Variables

```python
x = "awesome"

def myfunc():
  print("Python is " + x)

myfunc()
```

### Analisis

Kode tersebut menetapkan nilai "awesome" ke dalam variabel x dan membuat fungsi myfunc() yang mencetak teks hasil penggabungan dengan nilai variabel tersebut. Saat myfunc() dijalankan, program akan menampilkan output "Python is awesome" di layar.

```python
x = "awesome"

def myfunc():
  x = "fantastic"
  print("Python is " + x)

myfunc()

print("Python is " + x)
```

### Analisis

Kode tersebut mendefinisikan variabel global x = "awesome" dan variabel lokal x = "fantastic" di dalam fungsi myfunc(). Saat fungsi dijalankan, program menampilkan "Python is fantastic" karena menggunakan variabel lokal. Setelah fungsi selesai, perintah di luar fungsi mencetak "Python is awesome", yang berasal dari variabel global.

## The global Keyword

```python
def myfunc():
  global x
  x = "fantastic"

myfunc()

print("Python is " + x)
```

### Analisis

Kode tersebut menggunakan kata kunci global agar variabel x dapat diakses dan diubah di luar fungsi. Setelah myfunc() dijalankan, nilai variabel x menjadi "fantastic", sehingga output yang ditampilkan adalah "Python is fantastic".

```python
x = "awesome"

def myfunc():
  global x
  x = "fantastic"

myfunc()

print("Python is " + x)
```

### Analisis

Kode tersebut mula-mula mendefinisikan variabel x = "awesome", lalu di dalam fungsi myfunc() digunakan kata kunci global untuk memungkinkan perubahan nilai x di luar fungsi. Setelah fungsi dijalankan, nilai x berubah menjadi "fantastic", sehingga output program yang dihasilkan adalah "Python is fantastic".

## Type Conversion

```python
x = 1    # int
y = 2.8  # float
z = 1j   # complex

#convert from int to float:
a = float(x)

#convert from float to int:
b = int(y)

#convert from int to complex:
c = complex(x)

print(a)
print(b)
print(c)

print(type(a))
print(type(b))
print(type(c))
```

### Analisis 

Kode tersebut menunjukkan proses konversi tipe data numerik. Variabel x, y, dan z masing-masing berisi nilai int, float, dan complex. Nilai x dikonversi menjadi float disimpan pada variabel a, menghasilkan nilai 1.0. Selanjutnya, nilai y dikonversi menjadi int disimpan pada variabel b, menghasilkan nilai 2 karena bagian desimalnya dihilangkan. Terakhir, nilai x dikonversi menjadi complex dan disimpan pada variabel c, menghasilkan (1+0j). Saat program dijalankan, output yang tampilkan menunjukkan hasil konversi beserta tipe datanya, yaitu float, int, dan complex.

## The Walrus Operator

```python
numbers = [1, 2, 3, 4, 5]
count = len(numbers)
if count > 3:
    print(f"List has {count} elements")

if (count := len(numbers)) > 3:
    print(f"List has {count} elements")
```

### Analisis

Kode ini menggunakan list numbers = [1, 2, 3, 4, 5] untuk menunjukkan dua cara menghitung jumlah elemen. Pertama, panjang list dihitung menggunakan len() dan disimpan ke variabel count. Kedua, digunakan walrus operator (:=) yang memungkinkan proses perhitungan dan penugasan dilakukan dalam satu langkah, seperti if (count := len(numbers)) > 3:. Kedua metode ini sama-sama menghasilkan output “List has 5 elements”, tetapi versi dengan walrus operator membuat kode lebih ringkas dan efisien.

## Multiple Levels of Nesting

### Analisis

Kode Python ini menggunakan struktur if bersarang (nested if) untuk menentukan status kelulusan berdasarkan tiga kriteria: nilai (score), kehadiran (attendance), dan pengumpulan tugas (submitted).
Nilai score bernilai 85, attendance bernilai 90, dan submitted bernilai True. Program pertama memeriksa apakah score memenuhi batas kelulusan. Karena kondisinya benar, program melanjutkan untuk memeriksa attendance. Nilai kehadiran juga memenuhi syarat, sehingga program masuk ke pengecekan terakhir, yaitu submitted. Karena semua kondisi terpenuhi, program mencetak “Pass with good standing” ke layar.

## The Python Match Statement

```python
day = 4
match day:
  case 1:
    print("Monday")
  case 2:
    print("Tuesday")
  case 3:
    print("Wednesday")
  case 4:
    print("Thursday")
  case 5:
    print("Friday")
  case 6:
    print("Saturday")
  case 7:
    print("Sunday")
```

### Analisis

Kode tersebut menggunakan struktur match-case (fitur baru di Python 3.10 ke atas) untuk mencocokkan nilai variabel day. Variabel day bernilai 4, sehingga program menemukan kecocokan pada case 4 dan menjalankan perintah di dalamnya, mencetak “Thursday” ke layar. Program ini berfungsi mirip dengan pernyataan switch-case di bahasa pemrograman lain, dan memudahkan penulisan percabangan yang memiliki banyak kondisi.

## Default Value

```python
day = 4
match day:
  case 6:
    print("Today is Saturday")
  case 7:
    print("Today is Sunday")
  case _:
    print("Looking forward to the Weekend")
```

### Analisis

Kode tersebut menggunakan struktur match-case untuk mencocokkan nilai variabel day. Nilai day adalah 4, sehingga tidak cocok dengan case 6 maupun case 7. Akibatnya, program mengeksekusi case _, yang berfungsi sebagai default case untuk menangani semua nilai yang tidak sesuai dengan case lain, dan mencetak “Looking forward to the Weekend” ke layar. Kode ini menunjukkan cara menggunakan match-case dengan simbol _ untuk menangani kondisi umum atau nilai yang tidak tercakup secara spesifik.

## Combine Values

```python
day = 4
match day:
  case 1 | 2 | 3 | 4 | 5:
    print("Today is a weekday")
  case 6 | 7:
    print("I love weekends!")
```

### Analisis

Kode tersebut menggunakan struktur match-case untuk mengevaluasi nilai variabel day dan menampilkan pesan sesuai kategorinya. Variabel day bernilai 4, sehingga cocok dengan pola case 1 | 2 | 3 | 4 | 5, yang mewakili hari kerja (weekday). Akibatnya, program mengeksekusi blok tersebut dan mencetak “Today is a weekday” ke layar. Kode ini menunjukkan bagaimana beberapa nilai dapat digabungkan dalam satu case menggunakan tanda pemisah | (pipe) untuk membuat kondisi lebih ringkas dan mudah dibaca.

## If Statements as Guards

```python
month = 5
day = 4
match day:
  case 1 | 2 | 3 | 4 | 5 if month == 4:
    print("A weekday in April")
  case 1 | 2 | 3 | 4 | 5 if month == 5:
    print("A weekday in May")
  case _:
    print("No match")
```

### Analisis


Kode tersebut menggunakan struktur match-case dengan guard condition (if) untuk mencetak pesan berdasarkan nilai day dan month. Variabel month bernilai 5 dan day bernilai 4. Program pertama memeriksa case 1 | 2 | 3 | 4 | 5 if month == 4, tetapi kondisi month == 4 tidak terpenuhi sehingga dilewati. Selanjutnya, case 1 | 2 | 3 | 4 | 5 if month == 5 terpenuhi karena day cocok dan month sama dengan 5, sehingga program mencetak “A weekday in May” ke layar. Kode ini menunjukkan bagaimana guard condition digunakan dalam match-case untuk menambahkan syarat tambahan dan memperjelas logika pencocokan.
